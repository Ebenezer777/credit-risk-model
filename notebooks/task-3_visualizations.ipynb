```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")

# Load Task 3 features
df = pd.read_csv(
    r"C:\Users\assef\Desktop\Kifiya AI Mastery\week4\credit-risk-model\data\processed\task3_features.csv"
)

df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['total_amount'], bins=50, kde=True)
plt.title("Distribution of Total Transaction Amount per Customer")
plt.xlabel("Total Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(
    x='transaction_count',
    y='avg_amount',
    data=df,
    alpha=0.6
)
plt.title("Average Transaction Amount vs Transaction Count")
plt.xlabel("Transaction Count")
plt.ylabel("Average Amount")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['transaction_count'], bins=40, kde=True)
plt.title("Distribution of Transaction Counts per Customer")
plt.xlabel("Transaction Count")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['avg_tx_hour'], bins=24)
plt.title("Average Transaction Hour Distribution")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(x='common_tx_month', data=df)
plt.title("Most Common Transaction Month per Customer")
plt.xlabel("Month")
plt.ylabel("Customer Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df['channel'].value_counts().plot(kind='bar')
plt.title("Dominant Transaction Channel per Customer")
plt.xlabel("Channel")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df['product_category'].value_counts().plot(kind='bar')
plt.title("Dominant Product Category per Customer")
plt.xlabel("Product Category")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(10, 8))

numeric_df = df.select_dtypes(include='number')

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    square=True
)

plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['total_amount'])
plt.title("Outlier Detection  Total Amount")
plt.show()
```
